```python
import pandas as pd
```

```python
df = pd.read_csv('./assets/estudo_orientado_unidade_tres.csv').set_index('id')
```

```python
df.head()
```

```python
# Tratando dados faltantes
df['nome'].fillna('Faltante', inplace=True)
df['FrequenciaAtividadeFisica'].fillna(method='ffill', inplace=True)
df['Sexo'].fillna(method='bfill', inplace=True)
df.dropna(subset=['idade'], inplace=True)
```

```python
df.isna().sum()
```

```python
porcentagem = lambda x: (x['nome'] / df.shape[0]) * 100
df_porcentagem = df.groupby('FrequenciaAtividadeFisica').count().reset_index()
df_porcentagem = df_porcentagem.assign(Porcentagem=porcentagem)
df_porcentagem.round(2)
```

1- Qual a porcentagem de pessoas que praticam atividade física em cada uma das categorias: diariamente, semanalmente, raramente e não prática.

- diariamente: 13,98%
- semanalmente: 12,98%
- raramente: 10,98%
- não prática: 11,43%

2- Qual a porcentagem de homens e de mulheres que praticam atividade física pelo menos 1 vez por semana?

- Homens: 5,82% do total
- Mulheres:  5,71% do total

```python
df_semanalmente = df[((df['Sexo'] == 'Male') | (df['Sexo'] == 'Female')) & (df['FrequenciaAtividadeFisica'] == 'Weekly')]
df_semanalmente = df_semanalmente.groupby('Sexo').count().reset_index()
df_semanalmente = df_semanalmente.assign(Porcentagem=porcentagem).round(2)
df_semanalmente
```

3- Na faixa etária entre 20 e 29 anos, quantos homens e quantas mulheres praticam atividade física pelo menos 1 vez por semana? E quantos não praticam atividade física?

- 1 vez por semana: 29 pessoas
- Não praticam: 20 pessoas

```python
df_idade_vinte_vintenove = df[(df['idade'] >= 20) & (df['idade'] <= 29)].reset_index()
df_idade_vinte_vintenove.groupby('FrequenciaAtividadeFisica').count()
```
