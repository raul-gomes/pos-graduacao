```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = {
    'Meses': ['Abril', 'Maio', 'Junho', 'Julho'],
    'Total': [176543, 189021, 198765, 182123]
}

df = pd.DataFrame(dataset)
df
```

```python
tick_spacing = 3

fig, ax = plt.subplots(1,1)

fig.set_figheight(8)
fig.set_figwidth(20)

ax.fill_between(x=df['Meses'], y1=df['Total'], color='skyblue', alpha=0.4)
ax.plot(df['Meses'], df['Total'], color='Slateblue',  marker='o')
ax.set_ylim(50000, 250000)
ax.set_title('Total de Perder no período de Abril a Julho')
ax.set_xlabel('Meses')
ax.set_ylabel('Valor em perdas')
ax.axes.set_axisbelow(True)
ax.grid(True, color='gray', linestyle=':', linewidth=1)



fig.tight_layout()
```
