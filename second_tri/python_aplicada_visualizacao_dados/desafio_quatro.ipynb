```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Load DataFrame
df = pd.read_csv('./assets/imigrantes_canada-2.csv')

df.head()
```

```python
df_melted = df.melt(id_vars=['Continente', 'País', 'Região', 'Total'], var_name='Ano', value_name='Imigrantes')
df_brasil = df_melted[df_melted['País'] == 'Brasil']
df_brasil.head()
```

```python
df_melted = df.melt(id_vars=['Continente', 'País', 'Região', 'Total'], var_name='Ano', value_name='Imigrantes')
df_argelia = df_melted[df_melted['País'] == 'Argélia']
df_argelia.head()
```

```python
plt.figure(figsize=(20,8))
plt.fill_between(x=df_brasil['Ano'], y1=df_brasil['Imigrantes'], color='skyblue', alpha=0.4)
plt.plot(df_brasil['Ano'], df_brasil['Imigrantes'], color='Slateblue', alpha=0.6)
plt.show()

```

3. Explore a função MultipleLocator() e ajuste a frequência dos dados do eixo X;


```python
tick_spacing = 3

fig, ax = plt.subplots(1,1)

fig.set_figheight(8)
fig.set_figwidth(20)

ax.fill_between(x=df_brasil['Ano'], y1=df_brasil['Imigrantes'], color='skyblue', alpha=0.4)
ax.plot(df_brasil['Ano'], df_brasil['Imigrantes'], color='Slateblue', alpha=0.6)

ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

ax.set_title('Imigrantes saindo do Brasil para o Canadá')

ax.set_xlabel('Anos')
ax.set_ylabel('Quantidade de imigrantes')

fig.tight_layout()



```
