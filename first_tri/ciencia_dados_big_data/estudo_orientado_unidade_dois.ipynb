```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('assets/estudo_orientado_unidade_dois.csv', sep=';')

df
```

```python
df.isnull().sum()
```

```python
df.loc[1:1, ('Total de Doentes')] = int(df.loc[1:1, ('Total de curados')] + (df.loc[1:1, ('Total de curados')] * 0.54))
df.loc[3:3, ('Total de curados')] = int(df.loc[3:3, ('Total de Doentes')] - (df.loc[3:3, ('Total de Doentes')] * 0.46))
df.loc[4:4, ('Total de curados')] = int(df.loc[4:4, ('Total de Doentes')] - (df.loc[4:4, ('Total de Doentes')] * 0.46))
df
```

```python
df.loc[0:0, ('Total de Pessoas que fizeram o tratamento')] = int(df.loc[1:1, ('Total de Pessoas que fizeram o tratamento')]) * int(df.loc[0:0, ('Total de curados')]) // int(df.loc[1:1, ('Total de curados')])
df.loc[4:4, ('Total de Pessoas que fizeram o tratamento')] = int(df.loc[4:4, ('Total de curados')]) * int(df.loc[3:3, ('Total de Pessoas que fizeram o tratamento')]) // int(df.loc[3:3, ('Total de curados')])
```
