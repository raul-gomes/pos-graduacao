Função para realizar a leitura dos arquivos e retornar uma lista com os dados

```python
def ler_arquivo(path):

    with open(path,'r',newline='') as arq:
        # Efetuar a leitura dos dados no arquivo
        d = csv.reader(arq)
    
        # CONVERTER OS VALORES LIDOS EM UM NUMERO FLOAT
        file=[float(num[0]) for num in list(d)]
    
    return file
```

Função para plotar os datasets

```python
def plot_grafico(df_um, df_dois):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2 , figsize=(15,5), layout='tight')

    fig.suptitle('Comparação ')
    ax1.plot(df_um, 'r', linewidth=2) 
    ax1.set_title("Sinal Padrão")
    ax1.set_xlabel("Amostra[s]")
    ax1.set_ylabel("Amplitude")
    ax2.plot(df_dois, 'g', linewidth=2) 
    ax2.set_title("Sinal Dados")
    ax2.set_xlabel("Amostra[s]")
    ax2.set_ylabel("Amplitude")
  
```

Função para calcular a correlação dos dois datasets

```python
def correlacao(df_um, df_dois):
    auxDF = [num for num in zip(df_um, df_dois)]
    df = pd.DataFrame(auxDF, columns=['Padrão', 'Dados Amostra'])
    correlacao = df.corr(method='pearson')
    display(correlacao)

    print("Dados estástisticos básico")
    print("Numero de amostra por campo: ", df.shape)

    display(df.describe())
```

Comparação entres os datasets Padrão x Bigdata

```python
"""
Será feito uma quebra do arquvivo big data em 390 lista de 256 amostras cada com a função np.array_split.
Após essa quebra, será realizada a leitura individual de cada indice da lista comparando o padrão dentro de um loop while
caso o usuário deseja encerrar antes de finalizar a leitura, ele poderá realizar a qualquer momento.
"""

big_data = np.array_split(dados_big_data, 390)

count = 0

while count < len(big_data):
    plot_grafico(padrao, big_data[count])

    # MONTAGEM

    print("Correlação")

    correlacao(padrao, big_data[count])

    next = input("Deseja continuar [s-sim / n-não]: ")

    match next:
        case 's':
            count += 1
            continue
        case 'n':
            break
```
