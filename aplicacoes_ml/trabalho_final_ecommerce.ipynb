```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python


path_padrao = "assets_ecommerce/ecomercePadrao.txt"
path_janeiro = "assets_ecommerce/ecomerceJaneiro.txt"
path_fevereiro = "assets_ecommerce/ecomerceFevereiro.txt"
path_marco = "assets_ecommerce/ecomerceMarco.txt"
path_abril = "assets_ecommerce/ecomerceAbril.txt"
path_maio = "assets_ecommerce/ecomerceMaio.txt"
path_junho = "assets_ecommerce/ecomerceJunho.txt"
path_julho = "assets_ecommerce/ecomerceJulho.txt"
path_agosto = "assets_ecommerce/ecomerceAgosto.txt"
path_setembro = "assets_ecommerce/ecomerceSetembro.txt"
path_outubro = "assets_ecommerce/ecomerceOutubro.txt"
path_novembro = "assets_ecommerce/ecomerceNovembro.txt"
path_dezembro = "assets_ecommerce/ecomerceDezembro.txt"
```

```python
def ler_arquivo(path):

    with open(path,'r',newline='') as arq:
        # Efetuar a leitura dos dados no arquivo
        d = csv.reader(arq)
    
        # CONVERTER OS VALORES LIDOS EM UM NUMERO FLOAT
        file=[float(num[0]) for num in list(d)]
    
    return file
```

```python
def plot_grafico(df_um, df_dois):
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2 , figsize=(15,5), layout='tight')

    fig.suptitle('Comparação ')
    ax1.plot(df_um, 'r', linewidth=2) 
    ax1.set_title("Sinal Padrão")
    ax1.set_xlabel("Amostra[s]")
    ax1.set_ylabel("Amplitude")
    ax2.plot(df_dois, 'g', linewidth=2) 
    ax2.set_title("Sinal Dados")
    ax2.set_xlabel("Amostra[s]")
    ax2.set_ylabel("Amplitude")
```

```python
def correlacao(df_um, df_dois):
    auxDF = [num for num in zip(df_um, df_dois)]
    df = pd.DataFrame(auxDF, columns=['Padrão', 'Dados Amostra'])
    correlacao = df.corr(method='pearson')
    display(correlacao)

    print("Dados estástisticos básico")
    print("Numero de amostra por campo: ", df.shape)

    display(df.describe())
```

```python
dados_padrao = ler_arquivo(path_padrao)
dados_janeiro = ler_arquivo(path_janeiro)
dados_fevereiro = ler_arquivo(path_fevereiro)
dados_marco = ler_arquivo(path_marco)
dados_abril = ler_arquivo(path_abril)
dados_maio = ler_arquivo(path_maio)
dados_junho = ler_arquivo(path_junho)
dados_julho = ler_arquivo(path_julho)
dados_agosto = ler_arquivo(path_agosto)
dados_setembro = ler_arquivo(path_setembro)
dados_outubro = ler_arquivo(path_outubro)
dados_novembro = ler_arquivo(path_novembro)
dados_dezembro = ler_arquivo(path_dezembro)
```

```python
daily_report_janeiro = np.array_split(dados_janeiro, 30)
daily_report_padrao = np.array_split(dados_padrao, 30)

plot_grafico(daily_report_padrao[0], daily_report_janeiro[0])

print("Correlação")
correlacao(daily_report_padrao[0], daily_report_janeiro[0])
```
