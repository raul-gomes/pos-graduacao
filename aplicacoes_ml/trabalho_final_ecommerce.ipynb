```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

Aplicar correlação para avaliar o desempenho de faturamento de um e-comerce cujos dados estão listados nos arquivos (mensais) disponibilizados a seguir. A Diretoria estabeleceu um padrão de faturamento (arquivoPadrão.txt) que deverá ser utilizado para aferir o desempenho da empresa. Os arquivos contém  300 amostras obtidas em 10 horas diárias por 30 dias no mês.

#Análise

Através da análise dos gráficos foi possivel notar, que a correlação impacta diretamente no valor monetário obtido naquele mês, quanto mais proximo de 1, mais o valor se aproxima do padrão.
Em muitos casos a média de valores davam parecidos porém variava no min e max, podemos relacionar isso ao valor das vendas diárias, que não houveram flutuações ao longo do mês.


```python
# Cria um dicionário com os paths dos datasets
paths = {
    'padrao': "assets_ecommerce/ecomercePadrao.txt",
    'janeiro': "assets_ecommerce/ecomerceJaneiro.txt",
    'fevereiro': "assets_ecommerce/ecomerceFevereiro.txt",
    'marco': "assets_ecommerce/ecomerceMarco.txt",
    'abril': "assets_ecommerce/ecomerceAbril.txt",
    'maio': "assets_ecommerce/ecomerceMaio.txt",
    'junho': "assets_ecommerce/ecomerceJunho.txt",
    'julho': "assets_ecommerce/ecomerceJulho.txt",
    'agosto': "assets_ecommerce/ecomerceAgosto.txt",
    'setembro': "assets_ecommerce/ecomerceSetembro.txt",
    'outubro': "assets_ecommerce/ecomerceOutubro.txt",
    'novembro': "assets_ecommerce/ecomerceNovembro.txt",
    'dezembro': "assets_ecommerce/ecomerceDezembro.txt",
}
```

```python
def ler_arquivo(path: str) -> list:
    """Recebe um path e retorna o conteudo de arquivo.

    Args:
        path (str): caminho do arquivo

    Returns:
        list: uma lista do dos valores dentro do arquivo
    """
    with open(path,'r',newline='') as arq:
        # Efetuar a leitura dos dados no arquivo
        d = csv.reader(arq)
    
        # CONVERTER OS VALORES LIDOS EM UM NUMERO FLOAT
        file=[float(num[0]) for num in list(d)]
    
    return file
```

```python
def plot_grafico(df_um: list, df_dois:list, month: str) -> None:
    """Recebendo dois conjuntos de dados, faz o plot do gráfico de linhas comparando as informações

    Args:
        df_um (list): primeiro conjunto de dados
        df_dois (list): segundo conjunto de dados
        month (str): o mês a ser comparado
    """
    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2 , figsize=(15,5), layout='tight')

    fig.suptitle('Comparação ')
    ax1.plot(df_um, 'r', linewidth=2) 
    ax1.set_title("Amostra de venda padrão")
    ax1.set_xlabel(f"Amostra[s]")
    ax1.set_ylabel("Amplitude")
    ax2.plot(df_dois, 'g', linewidth=2) 
    ax2.set_title(f"Amostra do mês {month}")
    ax2.set_xlabel("Amostra[s]")
    ax2.set_ylabel("Amplitude")
```

```python
def correlacao(df_um: list, df_dois: list, month: str) -> None:
    """Recebendo dois conjuntos de dados, faz a correlação das informações

    Args:
        df_um (list): primeiro conjunto de dados
        df_dois (list): segundo conjunto de dados
        month (str): o mês a ser comparado
    """
    auxDF = [num for num in zip(df_um, df_dois)]
    print(f"Correlação entre o arquivo Padrão e o arquivo do mês {month}")
    
    df = pd.DataFrame(auxDF, columns=['Padrão', f'Dados mês {month}'])
    
    correlacao = df.corr(method='pearson')
    
    display(correlacao)

    print("Dados estástisticos básico")
    print("Numero de amostra por campo: ", df.shape)

    display(df.describe())
```

```python
# Gera dois datasets, o dataset com os valores padrões e o dataset com os valores mensais
data = {month: ler_arquivo(path) for month, path in paths.items() if month != 'padrao'}
data_padrao = ler_arquivo(paths['padrao'])
```

```python
# Plota o gráfico e faz a correlação dos dados
for month, sales_data in data.items():
    plot_grafico(data_padrao, sales_data, month)
    correlacao(data_padrao, sales_data, month)
```

```python
daily_report_janeiro = np.array_split(data['janeiro'], 30)
daily_report_padrao = np.array_split(data['padrao'], 30)

plot_grafico(daily_report_padrao[0], daily_report_janeiro[0])

print("Correlação")
correlacao(daily_report_padrao[0], daily_report_janeiro[0])
```
